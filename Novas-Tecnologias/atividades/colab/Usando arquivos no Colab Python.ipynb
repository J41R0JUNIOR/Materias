Usando arquivos no Colab:

1. Faça o upload do arquivo no Colab usando a aba da esquerda.

2. Abra o arquivo de entrada no Colab usando o comando open com a opção read (r)

3. Leia o arquivo linha a linha com readline.

4. Para gravação abra o arquivo de saída com opção write (w).

5. Grave no arquivo de saída usando o comando write.

6. Feche o arquivo com o comando close.




```python
with open('teste1.txt', 'r') as f:
  n = int(f.readline())
  print(n)
  l = []
  for i in range(n):
    line = f.readline()
    l.append(line)
    print(line)
```

```python
with open('saida.txt', 'w') as f:
  for x in l:
    f.write(x)
  f.close()
```
